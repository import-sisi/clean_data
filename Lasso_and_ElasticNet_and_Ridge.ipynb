```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('all_tehran_eslahi.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
df.drop(['ad_type'], axis=1, inplace=True)
df.drop(['Deposit', 'Lease'], axis=1, inplace=True)
df.drop(['vadie', 'ejare'], axis=1, inplace=True)
#df.drop(['vadie2', 'ejare2'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, columns=['place','Advertiser','category'])
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
#df['id'] = range(1, len(df) + 1)
```

```python
X = df.drop('vadie2',axis=1)
y = df.vadie2
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import math

# define a function to generate a prediction and then compare the desired metrics
def predict_metrics(lr, x, y):
    pred = lr.predict(x)
    mae = mean_absolute_error(y, pred)
    mse = mean_squared_error(y, pred)
    rmse = math.sqrt(mse)
    r2 = r2_score(y, pred)
    return mae, mse, r2 , rmse
```

```python
from sklearn.model_selection import train_test_split
multi_train_x, multi_test_x, multi_train_y, multi_test_y = train_test_split(X, y, test_size =0.07,random_state=101)
```

```python
from sklearn.linear_model import ElasticNet, Lasso
from sklearn.linear_model import Ridge
```

```python
lasso2 = Lasso(alpha =0.0005, random_state=1)
lasso2.fit(multi_train_x, multi_train_y)
multi_train_mae, multi_train_mse, multi_train_r2, multi_train_rmse = predict_metrics(lasso2, multi_train_x, multi_train_y)
multi_test_mae, multi_test_mse, multi_test_r2, multi_test_rmse = predict_metrics(lasso2, multi_test_x, multi_test_y)
print('training mean error: ', multi_train_mae, 'training mse: ', multi_train_mse, 'training r2: ', multi_train_r2, 'training rmse: ', multi_train_rmse)
print('test mean error: ', multi_test_mae, 'test mse: ', multi_test_mse, 'test r2: ', multi_test_r2, 'test rmse: ', multi_test_rmse)
```

```python
from sklearn.metrics import explained_variance_score


decpredict = lasso2.predict(multi_test_x)
exp_dec = explained_variance_score(multi_test_y,decpredict)
print(lasso2.score(multi_test_x, multi_test_y),'% Prediction Accuracy')
print(exp_dec,'% explained_variance_score')
```
