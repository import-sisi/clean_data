```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('all_tehran_eslahi.csv')
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
sns.heatmap(df.corr(),annot = True)
```

```python
df.drop(['vadie', 'ejare'], axis=1, inplace=True)
```

```python
df.drop(['Deposit', 'Lease'], axis=1, inplace=True)
```

```python
df.drop(['ad_type'], axis=1, inplace=True)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['place','Advertiser','category'])
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import math

# define a function to generate a prediction and then compare the desired metrics
def predict_metrics(lr, x, y):
    pred = lr.predict(x)
    mae = mean_absolute_error(y, pred)
    mse = mean_squared_error(y, pred)
    rmse = math.sqrt(mse)
    r2 = r2_score(y, pred)
    return mae, mse, r2 , rmse
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('vadie2',axis=1)
y = df.vadie2
```

```python
multi_train_x, multi_test_x, multi_train_y, multi_test_y = train_test_split(X, y, test_size =0.07 ,random_state=101)
```

```python
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
```

```python


model = lgb.LGBMRegressor()
model.fit(multi_train_x, multi_train_y)
```

```python
multi_train_mae, multi_train_mse, multi_train_r2, multi_train_rmse = predict_metrics(model , multi_train_x, multi_train_y)
multi_test_mae, multi_test_mse, multi_test_r2, multi_test_rmse = predict_metrics(model , multi_test_x, multi_test_y)
print('training mean error: ', multi_train_mae, 'training mse: ', multi_train_mse, 'training r2: ', multi_train_r2, 'training rmse: ', multi_train_rmse)
print('test mean error: ', multi_test_mae, 'test mse: ', multi_test_mse, 'test r2: ', multi_test_r2, 'test rmse: ', multi_test_rmse)
```

```python
estimator = lgb.LGBMRegressor(num_leaves=31)

param_grid = {
    'learning_rate': [0.01, 0.1, 1],
    'n_estimators': [20,500, 2000,100]
}

gbm = GridSearchCV(estimator, param_grid, cv=7)
gbm.fit(multi_train_x, multi_train_y)

print('Best parameters found by grid search are:', gbm.best_params_)
```

```python
multi_train_mae, multi_train_mse, multi_train_r2, multi_train_rmse = predict_metrics(gbm , multi_train_x, multi_train_y)
multi_test_mae, multi_test_mse, multi_test_r2, multi_test_rmse = predict_metrics(gbm , multi_test_x, multi_test_y)
print('training mean error: ', multi_train_mae, 'training mse: ', multi_train_mse, 'training r2: ', multi_train_r2, 'training rmse: ', multi_train_rmse)
print('test mean error: ', multi_test_mae, 'test mse: ', multi_test_mse, 'test r2: ', multi_test_r2, 'test rmse: ', multi_test_rmse)
```
