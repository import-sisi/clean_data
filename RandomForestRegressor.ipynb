```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('all_tehran_eslahi.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
df.drop(['ad_type'], axis=1, inplace=True)
df.drop(['Deposit', 'Lease'], axis=1, inplace=True)
df.drop(['vadie', 'ejare'], axis=1, inplace=True)
#df.drop(['vadie2', 'ejare2'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['place','Advertiser','category'])
```

```python
df.head()
```

```python
df = df.dropna()
```

```python

df['id'] = range(1, len(df) + 1)
```

```python
X = df.drop('vadie2',axis=1)
y = df.vadie2
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import math

# define a function to generate a prediction and then compare the desired metrics
def predict_metrics(lr, x, y):
    pred = lr.predict(x)
    mae = mean_absolute_error(y, pred)
    mse = mean_squared_error(y, pred)
    rmse = math.sqrt(mse)
    r2 = r2_score(y, pred)
    return mae, mse, r2 , rmse
```

```python
# import machine learning library
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split( X, y, test_size = 0.07, random_state = 101)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regr_rf = RandomForestRegressor()
regr_rf.fit(X_train, y_train)
y_rf = regr_rf.predict(X_test)
```

```python
multi_train_mae, multi_train_mse, multi_train_r2, multi_train_rmse = predict_metrics(regr_rf , X_train,y_train)
multi_test_mae, multi_test_mse, multi_test_r2, multi_test_rmse = predict_metrics(regr_rf , X_test, y_test)
print('training mean error: ', multi_train_mae, 'training mse: ', multi_train_mse, 'training r2: ', multi_train_r2, 'training rmse: ', multi_train_rmse)
print('test mean error: ', multi_test_mae, 'test mse: ', multi_test_mse, 'test r2: ', multi_test_r2, 'test rmse: ', multi_test_rmse)
```

```python
regrrf = RandomForestRegressor(n_estimators=271)
regrrf.fit(X_train, y_train)
multi_train_mae, multi_train_mse, multi_train_r2, multi_train_rmse = predict_metrics(regrrf , X_train,y_train)
multi_test_mae, multi_test_mse, multi_test_r2, multi_test_rmse = predict_metrics(regrrf , X_test, y_test)
print('training mean error: ', multi_train_mae, 'training mse: ', multi_train_mse, 'training r2: ', multi_train_r2, 'training rmse: ', multi_train_rmse)
print('test mean error: ', multi_test_mae, 'test mse: ', multi_test_mse, 'test r2: ', multi_test_r2, 'test rmse: ', multi_test_rmse)
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
rfgs_parameters = {
    'n_estimators': [100,200,50,271],
    'max_depth'   : [1,8,12,16],
    #'max_features': [n for n in range(1, 8)],
    #"min_samples_split": [n for n in range(2, 8)],
    #"min_samples_leaf": [n for n in range(2, 8)],
    #"bootstrap": [True,False]
}
rfr_cv = GridSearchCV(RandomForestRegressor(), rfgs_parameters, cv=7, scoring= 'neg_mean_squared_error')
rfr_cv.fit(X_train, y_train)
print("RFR GridSearch score: "+str(rfr_cv.best_score_))
print("RFR GridSearch params: ")
print(rfr_cv.best_params_)
```
