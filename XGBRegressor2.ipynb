```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('all_tehran_eslahi.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
df.drop(['vadie', 'ejare'], axis=1, inplace=True)
```

```python
df.drop(['Deposit', 'Lease'], axis=1, inplace=True)
```

```python
df.drop(['ad_type'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, columns=['place','Advertiser','category'])
```

```python
df = df.dropna()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import math

# define a function to generate a prediction and then compare the desired metrics
def predict_metrics(lr, x, y):
    pred = lr.predict(x)
    mae = mean_absolute_error(y, pred)
    mse = mean_squared_error(y, pred)
    rmse = math.sqrt(mse)
    r2 = r2_score(y, pred)
    return mae, mse, r2 , rmse
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('vadie2',axis=1)
y = df.vadie2
```

```python
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size =0.07 ,random_state=101)
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
import xgboost as xgb
from sklearn.metrics import accuracy_score
```

```python
param={'min_child_weight':[1, 2, 3, 6 ,5],'max_depth':[2, 8, 12, 6, 16],'eta':[0.1, 0.05, 0.01]}
model2 = GridSearchCV(xgb.XGBRegressor(),param_grid =param,cv = 7,verbose = True)

model2.fit(X_train, y_train)
print("SVR GridSearch score: "+str(model2.best_score_))
print("SVR GridSearch params: ")
print(model2.best_params_)
```

```python
trainDMat = xgb.DMatrix(data = X_train, label = y_train)
testDMat = xgb.DMatrix(data = X_test, label = y_test)
params = {'eta': 0.1, 'max_depth': 2, 'min_child_weight': 2}
```

```python
xgbCV = xgb.cv(
    params = params,  num_boost_round = 500, nfold = 7,dtrain = trainDMat,  verbose_eval = True, early_stopping_rounds = 10
)
```

```python

num_boost_round = len(xgbCV)

params = {'eta': 0.1, 'max_depth': 2, 'min_child_weight': 2,'eval_metric' : 'rmse'}
xgbFinal = xgb.train(
    params = params, 
    dtrain = trainDMat, 
    num_boost_round = num_boost_round,
    evals = [(trainDMat, 'train'), 
             (testDMat, 'eval')]
)
```

```python
xgbFinal_train_preds = xgbFinal.predict(xgb.DMatrix(X_train))
xgbFinal_test_preds = xgbFinal.predict(xgb.DMatrix(X_test))
```

```python
print ("\nModel Report")
print ("MSE Train : %f" % mean_squared_error(y_train, xgbFinal_train_preds))
print ("MSE Test: %f" % mean_squared_error(y_test, xgbFinal_test_preds))
print ("RMSE Train: %f" % mean_squared_error(y_train, xgbFinal_train_preds)**0.5)
print ("RMSE Test: %f" % mean_squared_error(y_test, xgbFinal_test_preds)**0.5)
```

```python
model_xgb = xgb.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1,
                             random_state =7, nthread = -1)
model_xgb.fit(X_train, y_train)
multi_train_mae, multi_train_mse, multi_train_r2, multi_train_rmse = predict_metrics(model_xgb , X_train,y_train)
multi_test_mae, multi_test_mse, multi_test_r2, multi_test_rmse = predict_metrics(model_xgb , X_test, y_test)
print('training mean error: ', multi_train_mae, 'training mse: ', multi_train_mse, 'training r2: ', multi_train_r2, 'training rmse: ', multi_train_rmse)
print('test mean error: ', multi_test_mae, 'test mse: ', multi_test_mse, 'test r2: ', multi_test_r2, 'test rmse: ', multi_test_rmse)
```

```python
model_xgb2 = xgb.XGBRegressor( 
                             learning_rate=0.1, max_depth=3, 
                             min_child_weight=2)
```

```python
model_xgb2.fit(X_train, y_train)
multi_train_mae, multi_train_mse, multi_train_r2, multi_train_rmse = predict_metrics(model_xgb2 , X_train,y_train)
multi_test_mae, multi_test_mse, multi_test_r2, multi_test_rmse = predict_metrics(model_xgb2 , X_test, y_test)
print('training mean error: ', multi_train_mae, 'training mse: ', multi_train_mse, 'training r2: ', multi_train_r2, 'training rmse: ', multi_train_rmse)
print('test mean error: ', multi_test_mae, 'test mse: ', multi_test_mse, 'test r2: ', multi_test_r2, 'test rmse: ', multi_test_rmse)
```
