```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from unidecode import unidecode
%matplotlib inline
```

```python
df= df[(df['Deposit'] !='توافقی')|(df['Lease'] !='توافقی')]
```

```python
def ejare(col):
    
    if (col != -1):
        a =((int(col) / 1000000) * 30000 )
        
        return a
        
    else :
        a = int(col)
        return col
```

```python
def vadie(col):
    
    if (col != -1):
        a =((int(col) / 30000) * 1000000 )
        a = int(a)
        return a
        
    else :
        a = int(col)
        return col
```

```python
def integer_per(col):
    
    if col == 'توافقی':
        a = -1
        return a
    else :
        a = unidecode(col)
            
       
        return a
```

```python
df['place']= df['place'].apply(lambda Lease: Lease.replace('تهران',''))
```

```python
df['Lease']= df['Lease'].apply(lambda Lease: Lease.replace('تومان',''))
df['Lease']= df['Lease'].apply(lambda Lease: Lease.replace('٫',''))
df['Lease']= df['Lease'].apply(lambda Lease: Lease.replace('مجانی','0'))
df['Lease']= df['Lease'].apply(lambda Lease: Lease.replace('رایگان','0'))
df['Lease']= df['Lease'].apply(integer_per)
```

```python
df['vadie']= df['Lease'].apply(vadie)
```

```python
df['Deposit']= df['Deposit'].apply(lambda Deposit: Deposit.replace('تومان',''))
df['Deposit']= df['Deposit'].apply(lambda Deposit: Deposit.replace('مجانی','0'))
df['Deposit']= df['Deposit'].apply(lambda Deposit: Deposit.replace('رایگان','0'))
df['Deposit']= df['Deposit'].apply(lambda Deposit: Deposit.replace('٫',''))
df['Deposit']= df['Deposit'].apply(integer_per)
```

```python
df['ejare']= df['Deposit'].apply(ejare)
```

```python
df['year_of_construction']= df['year_of_construction'].apply(lambda Lease: Lease.replace('قبل از',''))
df['year_of_construction']= df['year_of_construction'].apply(integer_per)
```

```python
df['Area']= df['Area'].apply(integer_per)
```

```python
dmap = {'بدون اتاق' : 0 , 'یک' : 1, 'دو' : 2 , 'سه' : 3, 'چهار' : 4,  'پنج ' : 5, 'پنج یا بیشتر' : 5 , 'صفر' : 0 }
df['Room_number']= df['Room_number'].map(dmap)
df.Room_number = df.Room_number.astype(str)
```
